```python
import cv2
from matplotlib import pyplot as plt
import numpy
```

```python
# The imread() function reads an image file in the form of a NumPy array
panda_img = cv2.imread("pandas_image.jpeg")
```

```python
# Printing the individual color channels f an image using the split() method

b,g,r= cv2.split(panda_img)
plt.imshow(b)
```

```python
plt.imshow(r)
```

```python
# convert the color channels of the image from BGR to RGB

panda_img_temp= cv2.imread(r"pandas_image.jpeg")
```

```python
panda_img= cv2.cvtColor(panda_img_temp, cv2.COLOR_BGR2RGB)
plt.imshow(panda_img)
```

```python
# Saveing an image in different format
# imwrite()

panda_img= cv2.imread(r"pandas_image.jpeg")
cv2.imwrite(r"panda_opencv.png", panda_img)
```

```python
# Converting an image to grayscale
panda_img_temp= cv2.imread(r"pandas_image.jpeg")
panda_img= cv2.cvtColor(panda_img_temp, cv2.COLOR_BGR2GRAY)
plt.imshow(panda_img)
```

```python
# cropping an image
panda_image= cv2.imread(r"pandas_img.jpg")
y=100
x=100
h=200
w=200
crop_img= panda_img[y:y+h, x:x+w]
plt.imshow(crop_img)
```

```python
#  Resizing an image
panda_img_temp= cv2.imread(r"pandas_image.jpeg")
resized_image= cv2.resize(panda_img_temp, (100,50))
plt.imshow(resized_image)
```

```python
#Applying filters
# Blur filter- the blur filter smooths an image by replacing pixels with the average value of surrounding pixels

panda_img_temp= cv2.imread(r"pandas_image.jpeg")
panda_img= cv2.blur(panda_img_temp, (10,10))
plt.imshow(panda_img)
```

```python
# Edge detection filter
panda_img_temp= cv2.imread(r"pandas_image.jpeg")
panda_img= cv2.Canny(panda_img_temp, 80,200)
plt.imshow(panda_img)
```

```python
#Changing image orientation
from PIL import Image
panda_img= Image.open(r"pandas_image.jpeg")
panda_img_ud= panda_img.transpose(Image.ROTATE_180)
display(panda_img_ud)
```

```python
panda_img_ud= panda_img.transpose(Image.FLIP_LEFT_RIGHT)
display(panda_img_ud)
```
